# Exploring IMF Data On Historical Public Debt

The aim of this project is to analyse the IMF's data on historical public debt to find out what trends can be found using this metric.

IMF Data: The Historical Public Debt Database contains unbalanced panel data on Gross Domestic Product, Gross Govenment Debt, and Gross Government Debt-to-GDP Ratio for 187 countries. The series spans the years 1800 through 2015 although each country's data depends on its date of independence and data availability. The database was constructed by bringing together a number of other datasets and information from original sources. For the most recent years, the data are linked to the IMF World Economic Outlook (WEO) database to facilitate regular updates.

### What Is the Debt-to-GDP Ratio?

The debt-to-GDP ratio is the metric comparing a country's public debt to its gross domestic product (GDP). By comparing what a country owes with what it produces, the debt-to-GDP ratio reliably indicates that particular country’s ability to pay back its debts. Often expressed as a percentage, this ratio can also be interpreted as the number of years needed to pay back debt, if GDP is dedicated entirely to debt repayment.

A country able to continue paying interest on its debt--without refinancing, and without hampering economic growth, is generally considered to be stable. A country with a high debt-to-GDP ratio typically has trouble paying off external debts (also called “public debts”), which are any balances owed to outside lenders. In such scenarios, creditors are apt to seek higher interest rates when lending. Extravagantly high debt-to-GDP ratios may deter creditors from lending money altogether.

Debt to GDP = Total Debt of Country / Total GDP of Country

### What Does the Debt-to-GDP Ratio Tell You? 

When a country defaults on its debt, it often triggers financial panic in domestic and international markets alike. As a rule, the higher a country’s debt-to-GDP ratio climbs, the higher its risk of default becomes. Although governments strive to lower their debt-to-GDP ratios, this can be difficult to achieve during periods of unrest, such as wartime, or economic recession. In such challenging climates, governments tend to increase borrowing in an effort to stimulate growth and boost aggregate demand.

A study by the World Bank found that countries whose debt-to-GDP ratios exceeds 77% for prolonged periods, experience significant slowdowns in economic growth. Pointedly: every percentage point of debt above this level costs countries 1.7% in economic growth. This phenomenon is even more pronounced in emerging markets, where each additional percentage point of debt over 64%, annually slows growth by 2%.

## Cleaning The Data

We can see that the dataset shows the Debt to GDP ratio for each country. In order for us to plot the data, we will need to transpose the dataset so that we have the years as a row with each country as a column. For the purposes of this analysis, we will not need the country code, indicator name, indicator code of attribute columns. These will therfore be removed to leave us with the Debt/GDP ratio per country.

```python
#Transpose data so it can be visualised. Used transpose() but .T can also be used
debt_data_trans = debt_data.set_index('Country Name').T.rename_axis('Year').rename_axis(None,axis=1).reset_index()
debt_data_trans.head(10)
```

```python
#Removing rows 1 - 5 so we just have the years data
debt_gdp = debt_data_trans.drop([0,1,2,3,220])
debt_gdp.head()
```

The data values were converted to strings when the dataset was transposed. We therefore need to convert the values back to numeric.

```python
cols = list(debt_gdp.columns)
debt_gdp = debt_gdp[cols].apply(pd.to_numeric)
print(debt_gdp.dtypes)
```

## Initial Analysis Of The Dataset

We can see that very few countries have data back to 1800 so we have decided to look at the debt/GDP ratio for countries starting with 1900.


```python
debt_gdp = debt_gdp[debt_gdp['Year'].between(1900,2015)]
debt_gdp['Year']
```

```python
debt_gdp.shape
```

```python
null_values = debt_gdp.isnull().sum().sort_index()
null_values
```

```python
null_values = debt_gdp.isnull().sum().sort_index()
plt.style.use('fivethirtyeight')
null_values.plot.hist()
plt.xlabel('Num. of Error Values')
plt.xlim(0,150)
plt.show()
```

We can see that very few countries have full dataset after 1900. Data after 1900 is a relevant sample size as it allows us to see how the debt to GDP ratio changed as a result of economic shocks such as war, famines and recession through the 20th century. The large number of null values gives us an opportunity to filter our dataset as it currently too large for any meaningful cross country analysis.

We now have a dataset looking at 116 years of data for each country. We have decided to remove countries with error counts of over 15. Given the type of data we are looking at, with few large differences between adjacent years, we can fill in the missing data using the Pandas interpolate method as this analysis is concerned with general trends as opposed to statistical analysis.

```python
debt_gdp = debt_gdp.dropna(axis=1, thresh=100)
debt_gdp.shape
debt_gdp.shape
```

```python
debt_gdp
```

```python
cols = list(debt_gdp.columns)
cols
```

```python
countries = [
 'Portugal',
 'Canada',
 'Norway',
 'New Zealand',
 'Sweden',
 'Italy',
 'South Africa',
 'Spain',
 'Netherlands',
 'Japan',
 'United Kingdom',
 'United States',
 'Argentina',
 'Australia',
 'Denmark']
```

```python
plt.style.use('fivethirtyeight')
colours = ['red','blue','green', 'yellow', 'black', 'brown', 'orange', 'purple', 'navy', 'violet', 'teal', 'green', 'black', 'gray', 'yellow', 'plum']
fig = plt.figure(figsize=(25,15))
plt.title("Debt/GDP Ratio From 1900 to 2015", y = 1.07)

for sp in range(0,15):
    ax = fig.add_subplot(5,3,sp+1)
    ax.plot(debt_gdp['Year'], debt_gdp[countries[sp]], color=colours[sp], label=countries[sp])
    ax.set_title(countries[sp])
    ax.set_xlim(1900,2015)
    if (sp == 0) or (sp == 3) or (sp == 6) or (sp == 9) or (sp == 12):
        plt.ylabel('Debt/GDP (%)')
    if (sp == 12) or (sp == 13) or (sp == 14):
        plt.xlabel('Year')
   
fig.tight_layout()

plt.show()
```

## Filling In Missing Data

We will use the pandas.DataFrame.interpolate method to fill in out missing values. We will start with the USA data which has a small number of missing values and thus can be easily replaced for the purpose of our analysis.

### USA

```python
countries_amend = [
 'Portugal',
 'Canada',
 'Norway',
 'New Zealand',
 'Sweden',
 'Italy',
 'Spain',
 'Netherlands',
 'Japan',
 'United Kingdom',
 'United States',
 'Argentina',
 'Australia',
 'Denmark']

for country in countries_amend:
    debt_gdp[country].interpolate(method='quadratic', inplace=True)
    
```

We can see that the data for our chosen countries is now complete. The clearest trends from our 15 countries this increase and following decrease in the debt/GDP ratio in the middle of the 20th century which reflects the 2nd world and the huge increases in national spending between 1939 and 1945. The following reduction in debt/gbp is more gradual for countries which were considered victorious (UK, USA, Netherlands, Canada etc.) as opposed to the countries who were defeated where the debt/GDP steeply reduces following the conclusion of the second world war. It would be expected that this is a result of a steep reduction in GDP given the damage that was done to these economies.

## Analysis Of The 4 Largest Economies

```python
plt.style.use('fivethirtyeight')
fig = plt.figure(figsize=(10,10))
plt.plot(debt_gdp['Year'], debt_gdp['United Kingdom'], label = 'UK', color = 'teal')
plt.plot(debt_gdp['Year'], debt_gdp['United States'], label = 'USA', color = 'green')
plt.plot(debt_gdp['Year'], debt_gdp['Japan'], label = 'Japan', color = 'pink')
plt.plot(debt_gdp['Year'], debt_gdp['Canada'], label = 'Canada', color = 'blue')
plt.legend(loc='upper right')
plt.ylabel('Debt/GDP (%)')
plt.show()
```

We can see from the our chart that clear moments in economic history can be identified from the ratios, namely:

- 1st World War:
    This can be seen by the sharp increase in debt between 1914 - 1920s
- 2nd World War:
    All countries in our chart were in some way involved in the second world war. We can even see the moments when they entered the war with a delay between the Uks increase in debt in comparison to the United States.
- The Post War Boom:
    The long decreases in the debit/GDP ratio are a reflection of the post war expansion as a result of sustained high growth together with full employment.
- 2007-08 Financial Crash:
    All countries experienced a sharp increase between 2007-08 which is a result of a deep recession, bank bailouts and a slow recovery of their economies. This is with the exception of Japan which experienced "the lost decade" from 1991 - 2000 and was later to become known as "the lost score"
